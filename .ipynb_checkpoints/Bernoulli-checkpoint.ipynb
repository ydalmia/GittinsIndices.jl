```julia
using GittinsIndices
using Bijections
using Plots
```

```julia
# See tables in the book: Multi-armed Bandit Allocation Indices
# this calculation matches the official published values!
function make_mab(trials, prior_p, prior_q)
    outcomes = binomial(trials, 2)
    T = zeros(outcomes, outcomes)
    R = zeros(outcomes)

    i = 1
    i_pq = Bijection()
    for p in prior_p : prior_p + trials
        for q in prior_q : prior_q + trials
            if (p + q) <= trials
                i_pq[i] = (p, q)
                i += 1
            end
        end
    end
    
    pq_i = inv(i_pq)

    for s in 1 : i - 1
        p, q = i_pq[s]
        if (p + 1, q) in keys(pq_i)
            ss = pq_i[(p + 1, q)]
            T[s, ss] = p / (p + q)
        end

        if (p, q + 1) in keys(pq_i)
            sf = pq_i[(p, q + 1)]
            T[s, sf] = q / (p + q)
        end

        R[s] = p / (p + q)
    end
    return T, R
end
```

```julia
gis = []
horizons = 25:5:50
alpha = 16
beta = 5
for horizon in horizons
    T, R = make_mab(horizon, alpha, beta)
    gi = calculate_discrete_gittins(
        length(R),
        T,
        R, 
        1,
        0.7,
    )
    push!(gis, gi)
end
```

```julia
plot(
    horizons, 
    abs.(gis .- 0.7743), 
    label="Gittins Index Error", 
    title="Error Against Gittins Indices from Book",
    xlabel="Horizon",
    ylabel="Absolute Error",
)
```
